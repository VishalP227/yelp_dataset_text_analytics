STATISTICAL TESTING ON UNLABELLED DATASET

During the process of Labelling we saw many mentions that were given by females, like they were stating 'with husband' or 'with boy friend'..... So what we did here found the total no of 'husband_mentions', and total no of 'wife_mentions' and then calculated the proportions of these values with the total sample size and performed a z-test(since the sample size was large) to prove our assumption statistically.



NULL HYPOTHESIS: The proportion of 'husband' mentions is lower than or equal to that of the proportion of 'wife' mentions.

ALTERNATIVE HYPOTHESIS: The proportion of 'husband' mentions is higher than the proportion of 'wife' mentions.

```python
# Assuming 'Review' is the column containing the reviews
reviews = final_df['text'].astype(str)
stat_df=pd.DataFrame()
stat_df['Review'] = final_df['text']
# Define the terms to count
terms_to_count = ['hubby', 'husband','wife', 'boyfriend', 'girlfriend','my guy','my girl']

# Count occurrences of each term in the reviews
for term in terms_to_count:
    stat_df[f'{term}_mentions'] = reviews.str.lower().str.count(term.lower())  # case-insensitive matching

# Display the DataFrame with the new columns
print(stat_df[['Review'] + [f'{term}_mentions' for term in terms_to_count]])
```

```python


#Filter rows where at least one of the columns is non-zero
filtered_df = stat_df[((stat_df['husband_mentions'] + stat_df['hubby_mentions'] +
                        stat_df['boyfriend_mentions'] + stat_df['my guy_mentions']) > 0) |
                      ((stat_df['wife_mentions'] + stat_df['girlfriend_mentions'] +
                        stat_df['my girl_mentions']) > 0)]

# Perform the t-test on the filtered DataFrame
husband_mentions_filtered = filtered_df['husband_mentions'] + filtered_df['hubby_mentions'] + filtered_df['boyfriend_mentions'] + filtered_df['my guy_mentions']
wife_mentions_filtered = filtered_df['wife_mentions'] + filtered_df['girlfriend_mentions'] + filtered_df['my girl_mentions']


print(filtered_df.shape)
print(filtered_df.head())

# Create a DataFrame with counts
counts_df = pd.DataFrame({
    'Count(H)': [husband_mentions_filtered.sum()],
    'Count(non-H)': [len(husband_mentions_filtered) - husband_mentions_filtered.sum()],
    'Count(W)': [wife_mentions_filtered.sum()],
    'Count(non-W)': [len(wife_mentions_filtered) - wife_mentions_filtered.sum()]
})

# Display the counts table
print(counts_df)
```

```python
import scipy.stats as stats


# Given data
count_husband_mentions = 111192
count_non_husband_mentions = 25468
count_wife_mentions = 82668
count_non_wife_mentions = 53992
total_reviews = 1142194


# Proportions
p_husband = count_husband_mentions / total_reviews
p_wife = count_wife_mentions / total_reviews


# Sample sizes
n_husband = count_husband_mentions + count_non_husband_mentions
n_wife = count_wife_mentions + count_non_wife_mentions



# Pooled proportion
p_pooled = (n_husband * p_husband + n_wife * p_wife) / (n_husband + n_wife)


# Z statistic
z_statistic = (p_husband - p_wife) / ((p_pooled * (1 - p_pooled)) * ((1 / n_husband) + (1 / n_wife)))**0.5


# Calculate p-value
p_value = 1 - stats.norm.cdf(z_statistic)


# Assuming a significance level of 0.05
alpha = 0.01

print("p_value =  ",p_value)
print("z_statistic = ",z_statistic)

# Compare p-value to significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence that the proportion of 'husband' mentions is higher than the proportion of 'wife' mentions.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest a difference in proportions.")
```

So the results of the z_test help us to reject the NULL Hypothesis. And hence there is enough evidence that 'husband' mentions are more than 'wife' mentions. This mean that 'The number of reviews given by female customers is higher than the number of reviews given by male customers'. JUST ONE OF THE DEMOGRAPHIC INSIGHT WE FOUND
